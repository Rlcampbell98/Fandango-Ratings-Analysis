In this project, we will analyze additional movie ratings from Fandango to determine if there have been any changes since the publication of Hickey's article.

This endeavor aims to leverage advanced statistical techniques and coding proficiency to scrutinize the integrity of Fandango's movie rating system, providing insights into its evolution since the publication of Walt Hickey's critical article.

#### Read and explore two datasets

```python
import pandas as pd

# load datasets
fandango_rating_before = pd.read_csv("fandango_score_comparison.csv")
fandango_rating_after = pd.read_csv("movie_ratings_16_17.csv")

# read and print out the first few line in both datasets
print("First Dataset: Fandango Score Comparison")
display(fandango_rating_before.head())

print("Second Dataset: Movie Rating 2016 - 2017")
display(fandango_rating_after.head())

# print an summary for both dataset
print("Summary of Fandango Score Comparison")
display(fandango_rating_before.info())

print("Summary of Movie Rating 2016 - 2017")
display(fandango_rating_after.info())

# print statistical summary for both dataset
print("Statistical Summary of Fandango Score Comparison")
display(fandango_rating_before.describe())

print("Statistical Summary of Movie Rating 2016 - 2017")
display(fandango_rating_after.describe())
```

```python
#  selecting certain column for this alanysis
fandango_previous = fandango_rating_before[['FILM','Fandango_Stars','Fandango_Ratingvalue','Fandango_votes','Fandango_Difference']]
fandango_after = fandango_rating_after[['movie','year','fandango']]

#read the variables that contain the dataset with selected columns to analyze
print("Dataset before Hickey's Analysis:")
display(fandango_previous.head())

print("Dataset after Hickey's Analysis:")
display(fandango_after.head())
```

### Define the population of interest and determine if sampling is random or not

#### The population of interest for this goal includes all movies available on Fandango's website with ratings posted both before and after Hickey's analysis. this sampling is not random because the movies included in the datasets were selected based on specific criteria, such as having at least 30 reviews and tickets available for sale. This approach doesn't provide an equal opportunity for all movies to be featured in the sample. As a result, it likely leaves out less popular or indie films that don't meet these thresholds, potentially skewing the data towards more mainstream or well-known movies.

### New Goal

#### Due to the non-random nature of the datasets and the challenges associated with analyzing the initial goal of detecting changes in ratings since Hickey's article, I have decided to modify my analytical approach. By focusing on the ratings of popular movies from 2015 and 2016, I aim to obtain a clearer understanding of the differences or similarities in the ratings over these years. This refined strategy will allow for a more targeted analysis of the most engaged films on Fandango's platform, thereby providing a more precise reflection of potential changes in the rating system.

```python
import pandas as pd
import matplotlib.pyplot as plt


# Activating the fivethirtyeight style
plt.style.use('fivethirtyeight')

# Create the figure and the axes
fig, ax = plt.subplots(figsize=(8, 6))

# Generate kernel density plots
fandango_previous['Fandango_Stars'].plot.kde(ax=ax, label='2015 Ratings')
fandango_after['fandango'].plot.kde(ax=ax, label='2016 Ratings')

# Setting the title and labels
ax.set_title('Distribution of Movie Ratings by Year', fontsize=16)
ax.set_xlabel('Rating')
ax.set_ylabel('Density')

# Setting the x-axis limits and ticks as specified
ax.set_xlim(0, 5)
ax.set_xticks([0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 3.5, 4.0, 4.5, 5.0])

# Adding a legend to differentiate between the years
ax.legend()

# Show the plot
plt.show()
```

### Analying the two kernel density plots

#### Looking at the kernel density plots for Fandango's movie ratings in 2015 and 2016, we can observe distinct patterns in how movies were rated each year. In 2015, the distribution is sharply skewed towards higher ratings, peaking around 4.5, which indicates a large number of movies received very high ratings. In contrast, the 2016 ratings show a broader distribution with a peak near 4.0, suggesting a more moderate rating pattern and a decrease in the concentration of top-end ratings. This shift suggests that while Fandango still favored higher ratings in 2016, there was a notable moderation compared to the previous year. The flatter and left-shifted peak in 2016 indicates a subtle yet clear reduction in average ratings, pointing towards a potential recalibration of Fandango's rating system following criticisms of its overly generous ratings practices highlighted in Hickey's analysis. This comparison hints that Fandango may have adjusted its system to provide a more balanced spread of movie ratings.

```python
import pandas as pd
import statistics

# Calculate mean, median, and mode for 2015
mean_2015 = fandango_previous['Fandango_Stars'].mean()
median_2015 = fandango_previous['Fandango_Stars'].median()
mode_2015 = statistics.mode(fandango_previous['Fandango_Stars'])

# Calculate mean, median, and mode for 2016
mean_2016 = fandango_after['fandango'].mean()
median_2016 = fandango_after['fandango'].median()
mode_2016 = statistics.mode(fandango_after['fandango'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data preparation
labels = ['Mean', 'Median', 'Mode']
stats_2015 = [mean_2015, median_2015, mode_2015]
stats_2016 = [mean_2016, median_2016, mode_2016]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, stats_2015, width, label='2015')
rects2 = ax.bar(x + width/2, stats_2016, width, label='2016')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Ratings')
ax.set_title('Comparison of Rating Statistics for 2015 and 2016')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```

### Analysis from the Plot

#### Analyzing the grouped bar plot comparing the rating statistics for Fandango in 2015 and 2016, we observe a few key differences. The mean rating in 2016 is slightly lower than in 2015, decreasing from approximately 4.09 to 3.89, indicating a subtle reduction in overall ratings. Interestingly, the median rating remains consistent at 4.0 across both years, suggesting that the central tendency of movie ratings did not undergo significant shifts. However, a notable change is seen in the mode, which decreased from 4.5 in 2015 to 4.0 in 2016. This shift suggests that fewer movies in 2016 achieved the highest frequent rating of 4.5, pointing to a more moderate rating pattern.

# Conclusion 

#### The direction and magnitude of these differences suggest that Fandango may have implemented slight adjustments to its rating system following the criticism it received. The reduction in the mean and mode, alongside the stability of the median, indicates a nuanced recalibration of the rating distribution, likely aimed at tempering the propensity to award the highest ratings. This strategic response appears designed to refine the rating practices without drastically altering the platform's user-friendly, high-rating reputation. By subtly adjusting ratings, Fandango could enhance the credibility of its review system while maintaining a high level of user satisfaction.
